# Superscore Model
### Imports/Setup
The [`SuperScoreModel`](../src/superscore.py) class is a wrapper around the [`Result`](../utils/results.py) class that allows for addition methods that "superscore" each teams' results as another metric to measure team strength. This proves to be a useful metric for predicting future results, as it is a more accurate representation of a team's theoretical strength than their "unstack" results.

The attributes for the `SuperScoreModel` class are as follows:
```py
class SuperScoreModel(Results):
    ...
    @property
    def super_scores(self) -> dict[str, int]:
        # Returns a dictionary of each team's superscore
        ...
        
```
All other logic is auto handled by the `Results` superclass.
Using formula: $\sum_k^n{\min{(s_k, t_k, u_k,...)}}$ where $k$ is the event, and $s, t, u$ are different teams of the same school

````python
from src.superscore import SuperScoreModel

model = SuperScoreModel(results_path="../data/2023-02-18_penn_invitational_c.yaml")
````

### View Super Score
Sorted in order

````python
for i, (team, score) in enumerate(model.super_scores.items()):
    print(f"{i+1}. {team}: {score}")
````
